```python
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# Загружаем датасет Iris
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Масштабируем данные
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Разделим данные на обучающую и тестовую выборки
X_train = np.concatenate((X[0:30], X[50:80], X[100:130]))
X_test = np.concatenate((X[30:50], X[80:100], X[130:150]))
y_train = np.concatenate((y[0:30], y[50:80], y[100:130]))
y_test = np.concatenate((y[30:50], y[80:100], y[130:150]))
```

```python
np.random.seed(42)

# === Параметры SNN ===
num_inputs = 4           # 4 признака у Iris
num_neurons = 3          # Один нейрон на каждый класс
time_steps = 20          # Эмуляция по времени
threshold = 1.0
decay = 0.9
epochs = 30

# === STDP параметры ===
A_plus = 0.02
A_minus = -0.024
tau_stdp = 5


# Спайковое кодирование: преобразуем значения признаков в вероятность спайка
def encode_input(x_sample):
    return np.random.rand(time_steps, num_inputs) < x_sample

# === Инициализация весов ===
weights = np.random.rand(num_inputs, num_neurons)

# === LIF-нейрон ===
def run_lif(inputs, weights):
    mem = np.zeros(num_neurons)
    spikes = np.zeros((time_steps, num_neurons))
    first_spike_times = np.full(num_inputs, np.inf)
    for t in range(time_steps):
        mem = decay * mem + np.dot(inputs[t], weights)
        fired = mem >= threshold
        spikes[t] = fired
        mem[fired] = 0
        for i, spike in enumerate(spikes[t]):
            if (spike and first_spike_times[i] > t):
                first_spike_times[i] = t

    return spikes, first_spike_times

# === STDP обновление ===
def stdp_update(pre_spikes, post_spikes, weights):
    for t in range(time_steps):
        for pre in range(num_inputs):
            for post in range(num_neurons):
                dt = np.where(post_spikes[:, post])[0] - t
                if pre_spikes[t, pre]:
                    for d in dt:
                        if -tau_stdp < d <= 0:
                            weights[pre, post] += A_plus * np.exp(d / tau_stdp)
                        elif 0 < d < tau_stdp:
                            weights[pre, post] += A_minus * np.exp(-d / tau_stdp)
    weights = np.clip(weights, 0, 1)
    return weights

# === Обучение ===
for epoch in range(epochs):
    for x, label in zip(X_train, y_train):
        encoded = encode_input(x)
        output_spikes, first_spike_times = run_lif(encoded, weights)
        target_spikes = np.zeros_like(output_spikes)
        target_spikes[10:, label] = 1  # "цель": спайки только нужного нейрона
        weights = stdp_update(encoded, target_spikes, weights)
    y_pred = [predict(x) for x in X_train]
    print(accuracy_score(y_train, y_pred))

# === Предсказание ===
def predict(x):
    encoded = encode_input(x)
    output_spikes, first_spike_times = run_lif(encoded, weights)
    #spike_counts = output_spikes.sum(axis=0)
    return np.argmin(first_spike_times)

def plot_activity(encoded_input, weights, label=None):
    mem_trace = np.zeros((time_steps, num_neurons))
    spikes = np.zeros((time_steps, num_neurons))
    mem = np.zeros(num_neurons)

    for t in range(time_steps):
        mem = decay * (mem + np.dot(encoded_input[t], weights))
        fired = mem >= threshold
        spikes[t] = fired
        mem_trace[t] = mem
        mem[fired] = 0

    fig, ax1 = plt.subplots(figsize=(10, 5))

    # Мембранный потенциал
    for i in range(num_neurons):
        ax1.plot(range(time_steps), mem_trace[:, i], label=f"Membrane N{i}")
    ax1.set_ylabel("Membrane Potential")
    ax1.set_xlabel("Time Step")
    ax1.legend(loc="upper left")

    # Спайки
    ax2 = ax1.twinx()
    for i in range(num_neurons):
        spike_times = np.where(spikes[:, i])[0]
        ax2.plot(spike_times, [i + 0.5] * len(spike_times), 'ko', markersize=5, label=f"Spikes N{i}")
    ax2.set_yticks([])
    ax2.set_ylabel("Spikes")

    if label is not None:
        plt.title(f"Наблюдение класса: {label}")
    plt.tight_layout()
    plt.show()

def plot_potential_history(X_batch, y_batch, weights):

    offset = 2.0  # расстояние между примерами по вертикали

    all_mem_trace = np.zeros((time_steps * len(X_batch), num_neurons))
    all_spikes = np.zeros((time_steps * len(X_batch), num_neurons))

    for idx in range(len(X_batch)):
        x = X_batch[idx]
        y_true = y_batch[idx]
        encoded = encode_input(x)

        mem = np.zeros(num_neurons)
        mem_trace = np.zeros((time_steps, num_neurons))
        spikes = np.zeros((time_steps, num_neurons))

        for t in range(time_steps):
            mem = decay * mem + np.dot(encoded[t], weights)
            fired = mem >= threshold
            all_spikes[time_steps * idx + t] = fired
            spikes[t] = fired
            all_mem_trace[time_steps * idx + t] = mem
            mem_trace[t] = mem
            mem[fired] = 0

    for i in range(num_neurons):
        plt.figure(figsize=(25, 10))
        plt.plot(all_mem_trace[:, i], label='Membrane potential', linewidth=1, zorder=1)

        # Отметим спайки крестиками
        spike_times = [t for t, spike in enumerate(all_spikes[:, i]) if spike]
        spike_values = [threshold] * len(spike_times)  # по уровню порога
        plt.scatter(spike_times, spike_values, color='red', marker='x', label='Spike', zorder=2)

        plt.xlabel('Time step')
        plt.ylabel('Membrane potential')
        plt.title(f'Output Neuron {i} Potential & Spikes')
        plt.axhline(threshold, color='gray', linestyle='--', label='Threshold')
        plt.ylim(min(all_mem_trace[:, i]), max(all_mem_trace[:, i]))

        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

y_pred = [predict(x) for x in X_test]
acc = accuracy_score(y_test, y_pred)
print("Точность на тесте:", acc)

plot_potential_history(X_test, y_test, weights)
```
